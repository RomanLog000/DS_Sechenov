# Outliers analysis

```python
# Function for checking outliers using interquartile range (IQR)
def detect_outliers(df, column):
    """
    Detects outliers in a numerical column using the Interquartile Range (IQR) method.

    Args:
        df (pd.DataFrame): The dataset containing the column to be analyzed.
        column (str): The name of the numerical column to check for outliers.

    Returns:
        pd.DataFrame: Subset of the dataset containing rows identified as outliers.

    Outlier Criteria:
        - Values below Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR.
    """
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers
```

```python
# Function for outlier analysis and boxplot construction
def analyze_outliers_and_plot(data, dataset_name):
    """
    Analyzes and visualizes outliers for numerical columns in a dataset.

    Args:
        data (pd.DataFrame): The dataset to be analyzed.
        dataset_name (str): A name or label for the dataset.

    Outputs:
        - Prints the count of outliers for each numerical column.
        - Displays boxplots for all numerical columns to visualize outliers.
    """
    print(f"\nAnalyzing dataset: {dataset_name}")
    numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns

    for col in numerical_columns:
        outliers = detect_outliers(data, col)
        print(f"Column '{col}' has {len(outliers)} outliers.")

    # Boxplots
    plt.figure(figsize=(15, 10))
    for i, col in enumerate(numerical_columns, 1):
        plt.subplot(len(numerical_columns) // 3 + 1, 3, i)
        sns.boxplot(y=data[col], color='skyblue')
        plt.title(f"Boxplot of {col}")
    plt.tight_layout()
    plt.show()


# Analysis for each dataset 
analyze_outliers_and_plot(nafld_1, "NAFLD1")
analyze_outliers_and_plot(nafld_2, "NAFLD2")
analyze_outliers_and_plot(nwtco, "NWTCO")
```

```python
# --- Annomalies clean up with duplicate removal ---

# Clearing nafld_1 of BMI values outside the 10-60 range and duplicates
def clean_nafld1_bmi(df):
    """
    Cleans a dataset by:
        1. Removing rows with BMI values outside the range of 10 to 60.
        2. Dropping duplicate rows.

    Args:
        df (pd.DataFrame): The dataset to be cleaned.

    Returns:
        pd.DataFrame: Cleaned dataset with BMI values in the valid range and duplicates removed.

    Outputs:
        - Number of records removed during the cleaning process.
    """
    initial_len = len(df)
    df = df[(df['bmi'] >= 10) & (df['bmi'] <= 60)]
    df = df.drop_duplicates().reset_index(drop=True)
    print(f"NAFLD1: {initial_len - len(df)} items dropped.")
    return df

# Clearing nafld_2 of anomalies and duplicates
def clean_nafld2_anomalies(df):
    """
    Cleans a dataset by:
        1. Removing rows with specific anomalies based on test values.
        2. Dropping duplicate rows.

    Args:
        df (pd.DataFrame): The dataset to be cleaned.

    Returns:
        pd.DataFrame: Cleaned dataset with anomalies and duplicates removed.

    Anomaly Criteria:
        - hdl > 120
        - sbp outside 90-180
        - dbp outside 60-120
        - fib4 > 4

    Outputs:
        - Number of records removed during the cleaning process.
    """
    initial_len = len(df)
    
    # High-density lipoprotein levels (hdl) >120
    df = df[~((df['test'] == 'hdl') & (df['value'] > 120))]
    
    # Systolic BP level  (sbp) 90-180
    df = df[~((df['test'] == 'sbp') & ((df['value'] < 90) | (df['value'] > 180)))]
    
    # Diastolic BP level (dbp) 60-120
    df = df[~((df['test'] == 'dbp') & ((df['value'] < 60) | (df['value'] > 120)))]
    
    # Liver  fibrosis level (fib4) >4
    df = df[~((df['test'] == 'fib4') & (df['value'] > 4))]
    
    # Duplicates dropping
    df = df.drop_duplicates().reset_index(drop=True)
    
    print(f"NAFLD2: {initial_len - len(df)} items dropped.")
    
    return df

# Clearing nwtco of age > 120 and duplicates
def clean_nwtco_age(df):
    """
    Cleans a dataset by:
        1. Removing rows with age > 120.
        2. Dropping duplicate rows.

    Args:
        df (pd.DataFrame): The dataset to be cleaned.

    Returns:
        pd.DataFrame: Cleaned dataset with valid age values and duplicates removed.

    Outputs:
        - Number of records removed during the cleaning process.
    """
    initial_len = len(df)
    df = df[df['age'] <= 120]
    df = df.drop_duplicates().reset_index(drop=True)
    print(f"NWTC: {initial_len - len(df)} itens dropped.")
    return df

# Applying function
nafld_1_cleaned = clean_nafld1_bmi(nafld_1)
nafld_2_cleaned = clean_nafld2_anomalies(nafld_2)
nwtco_cleaned = clean_nwtco_age(nwtco)
```

# NAFLD2 optimization

```python
# Creating DataFrames

# Create bp (sbp/dbp mean)
bp_data = nafld_2_cleaned[nafld_2_cleaned['test'].isin(['sbp', 'dbp'])]
bp_mean = bp_data.groupby(['id', 'days']).agg({'value': 'mean'}).reset_index()
bp_mean.rename(columns={'value': 'bp'}, inplace=True)

# Create ratio chol/hdl
chol_hdl_data = nafld_2_cleaned[nafld_2_cleaned['test'].isin(['chol', 'hdl'])]
chol_hdl_ratio = chol_hdl_data.pivot_table(index=['id', 'days'], columns='test', values='value').reset_index()
chol_hdl_ratio['chol / hdl'] = chol_hdl_ratio['chol'] / chol_hdl_ratio['hdl']
chol_hdl_ratio = chol_hdl_ratio[['id', 'days', 'chol / hdl']]

# Create fib4
fib4_data = nafld_2_cleaned[nafld_2_cleaned['test'] == 'fib4']
fib4_mean = fib4_data.groupby(['id', 'days']).agg({'value': 'first'}).reset_index()
fib4_mean.rename(columns={'value': 'fib4'}, inplace=True)

# Create smoke
smoke_data = nafld_2_cleaned[nafld_2_cleaned['test'] == 'smoke']
smoke_mean = smoke_data.groupby(['id', 'days']).agg({'value': 'first'}).reset_index()
smoke_mean.rename(columns={'value': 'smoke'}, inplace=True)

# Merge DaraFrames
final_result = bp_mean.merge(chol_hdl_ratio, on=['id', 'days'], how='outer')
final_result = final_result.merge(fib4_mean, on=['id', 'days'], how='outer')
final_result = final_result.merge(smoke_mean, on=['id', 'days'], how='outer')

# Sort by id and days
final_result.sort_values(by=['id', 'days'], ascending=[True, True], inplace=True)

nafld_2_cleaned_pivot = final_result
```

- We see that in all cases, the parameters chol and hdl are equal, so we will exclude one of these indicators (hdl) from the analysis.

- We combine the BP parameter from sbp/dbp, as they are also equal, but just in case, we take the average.

```python
# Extract items from nafld_2_cleaned
chol_data = nafld_2_cleaned[nafld_2_cleaned['test'] == 'chol']
chol_values = chol_data[['id', 'days', 'value']].rename(columns={'value': 'chol'})

# Merge items chol with nafld_2_cleaned_pivot
nafld_2_cleaned_pivot = nafld_2_cleaned_pivot.merge(chol_values, on=['id', 'days'], how='left')

# Delete previous column chol / hdl
nafld_2_cleaned_pivot.drop(columns=['chol / hdl'], inplace=True)

# Rename column 'chol'
nafld_2_cleaned_pivot.rename(columns={'chol': 'chol'}, inplace=True)
```

```python
# --- 1. Distribution graphs and categorical feature analysis ---

# Updating categorical features
nafld_union['sex'] = nafld_union['sex'].map({'M': 'Male', 'F': 'Female'})
nafld_union['status'] = nafld_union['status'].map({'A': 'Alive', 'D': 'Dead'})

# Updating columns' lists
numerical_columns = nafld_union.select_dtypes(include=['float64', 'int64']).columns.tolist()
categorical_columns = ['sex', 'status']
numerical_columns_block1 = ['age', 'weight', 'height', 'bmi', 'case.id', 'futime']
numerical_columns_block2 = [col for col in numerical_columns if col.startswith('bp') or col.startswith('chol') or col.startswith('fib4')]

# --- Plots blocking ---
def plot_block(columns, title, is_categorical=False):
    """
    Creates a grid of plots for given columns in a dataset.

    Args:
        columns (list): List of column names to be plotted.
        title (str): Title for the entire block of plots.
        is_categorical (bool): Indicates whether the columns are categorical.

    Outputs:
        - Displays a grid of countplots for categorical columns or histograms for numerical columns.
        - Automatically adjusts the grid layout based on the number of columns.
    """
    num_cols = 4  # Number of plots in a string
    num_rows = -(-len(columns) // num_cols)  # Calculate rows number
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, num_rows * 5))
    fig.suptitle(title, fontsize=20, y=1.02)

    for i, column in enumerate(columns):
        ax = axes[i // num_cols, i % num_cols] if num_rows > 1 else axes[i % num_cols]
        if is_categorical:
            if nafld_union[column].notnull().any():  # Check data
                sns.countplot(
                    data=nafld_union,
                    x=column,
                    ax=ax,
                    order=nafld_union[column].value_counts().index,
                    palette='pastel',
                    hue='sex',  # Укажите hue, если хотите разделить по полу
                    legend=False  # Если не хотите отображать легенду
                )
                ax.set_title(f"Category: {column}")
            else:
                ax.text(0.5, 0.5, f"No data in {column}", ha='center', va='center')
                ax.set_title(f"Category: {column}")
        else:
            sns.histplot(nafld_union[column], kde=True, bins=30, ax=ax, color='blue')
            ax.set_title(f"Distribution: {column}")
        ax.set_xlabel(column)
        ax.set_ylabel("Frequency")
    
    # Disabling empty sub-graphs
    for j in range(i + 1, num_rows * num_cols):
        fig.delaxes(axes[j // num_cols, j % num_cols] if num_rows > 1 else axes[j % num_cols])
    
    plt.tight_layout()
    plt.show()


# Block 1: Numerical features (demographic data)
plot_block(numerical_columns_block1, "Distributions: Block 1 (Demographics and Case Info)", is_categorical=False)

# Block 2: Categorical features 
plot_block(categorical_columns, "Distributions: Block 2 (Categorical Features)", is_categorical=True)    

# Block 3: Value-parameters
plot_block(numerical_columns_block2, "Distributions: Block 3 (Value Parameters)", is_categorical=False)

# --- 2. Building a heat map for age, height, weight, and median bp, fib4, and chol ---
median_columns = ['bp_median', 'fib4_median', 'chol_median']
demographics = ['age', 'height', 'weight']
heatmap_data = nafld_union[demographics + median_columns]

# Heatmap for males and femaales separately
for gender, group_data in nafld_union.groupby('sex'):
    plt.figure(figsize=(16, 14))
    correlation_matrix = group_data[heatmap_data.columns].corr()
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
    plt.title(f"Correlation Heatmap ({gender})")
    plt.show()

# --- 3. Plotting paired graphs for age and medians ---
pairplot_columns = ['age'] + median_columns

# Copy for visualization
visual_data = nafld_union.copy()
sns.pairplot(
    visual_data[pairplot_columns + ['sex']],
    diag_kind='kde',
    hue='sex',
    palette={'Male': 'blue', 'Female': 'red'},
    plot_kws={'alpha': 0.6}
)
plt.suptitle("Pairplot for Age and Median Parameters by Gender", y=1.02)
plt.show()
```
